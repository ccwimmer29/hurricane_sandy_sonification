```python
import csv
import numpy as np
from scipy.io.wavfile import write
import pandas as pd

df = pd.read_csv("hurr_sandy.csv")
hurr_data = df.iloc[:,2:7].values
new_df = pd.DataFrame(hurr_data)
new_df = new_df.replace('-', 0)

# wind speed [:,0], wind dir (deg)[:,1], wind gust[:,2], air temp[:,3], baro[:,4]
wind_speed = new_df.iloc[1:,0]
wind_speed.to_csv('wind_speed.csv', index=False)

wind_gust = new_df.iloc[1:,2]
wind_gust.to_csv('wind_gust.csv', index=False)

baro = new_df.iloc[1:,4]
baro.to_csv('baro.csv', index=False)

water_level = pd.read_csv("water_level.csv")
level_data = water_level.iloc[:477,4].values
level_df = pd.DataFrame(level_data).replace('-',0)
level_df.to_csv('level.csv', index=False)
df
```

```python
# Load CSV data
wind_speed_d = []
with open('wind_speed.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        wind_speed_d.append([float(cell) for cell in row])

wind_gust_d = []
with open('wind_gust.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        wind_gust_d.append([float(cell) for cell in row])

baro_d = []
with open('baro.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        baro_d.append([float(cell) for cell in row])

level_d = []
with open('level.csv', 'r') as csvfile:
    reader = csv.reader(csvfile)
    for row in reader:
        level_d.append([float(cell) for cell in row])
```

```python
import pretty_midi

# WIND SPEED
# Normalize data to fit within MIDI note range
max_val = max(max(row) for row in wind_speed_d)
min_val = min(min(row) for row in wind_speed_d)
range_val = max_val - min_val
data_norm = [[(val - min_val) / range_val for val in row] for row in wind_speed_d]

# Convert data to MIDI notes
midi_data = pretty_midi.PrettyMIDI()
piano_program = pretty_midi.instrument_name_to_program('Acoustic Grand Piano')
piano = pretty_midi.Instrument(program=piano_program)
time = 0
for row in data_norm:
    for val in row:
        note_number = int(val * 88) + 21  # Map normalized value to MIDI note range
        note = pretty_midi.Note(
            velocity=100,
            pitch=note_number,
            start=time,
            end=time+0.1
        )
        piano.notes.append(note)
        time += 0.1
midi_data.instruments.append(piano)

# Save MIDI file
midi_data.write('wind_speed.mid')
```
